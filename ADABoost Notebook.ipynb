```python
#Import necessary packages
import tensorflow
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load MNIST and FASHION_MNIST data
mnist = keras.datasets.mnist
fashion_mnist = keras.datasets.fashion_mnist

#The Keras datasets load in this weird way
(mnist_train_images, mnist_train_labels), (mnist_test_images, mnist_test_labels) = (
    mnist.load_data())
(fmnist_train_images, fmnist_train_labels), (fmnist_test_images, fmnist_test_labels) = (
    fashion_mnist.load_data())
```

```python
mnist_images = np.concatenate((mnist_train_images, mnist_test_images))
n_mnist_images = len(mnist_images)
mnist_labels = np.array([1] * n_mnist_images)

fashion_images = np.concatenate((fmnist_train_images, fmnist_test_images))
n_fashion_images = len(fashion_images)
fashion_labels = np.array([0] * n_fashion_images)
```

```python
class_names = ['fashion', 'number']
images = np.concatenate((mnist_images, fashion_images))
labels = np.concatenate((mnist_labels, fashion_labels))
```

```python
plotting_indices = np.random.randint(0,len(labels),25)
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(images[plotting_indices[i]], cmap=plt.cm.binary)
    plt.xlabel(class_names[labels[plotting_indices[i]]])
```
